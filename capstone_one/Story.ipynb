It shows that people are using their blurbs to describe their project, which is great. Finally, there are the unifying words such as "one", "life", "world", and "love" that just elicits feelings of oneness and bringing people together. That's essentially what a kickstarter and crowdfunding is. These creators are effectively using the blurb to get a prospective backer's attention.

Immediately, I'd like to see whether or not many of these variables are connected to our eng goal which is the 'state' of the project.   
Alternatively, we can create a 'success' column on whether the percent goal reached 100% or not. 

```python
df['reach_goal'] = df['percent_goal'] >= 100.00
```

```python
fig, ax = plt.subplots(figsize=(4, 6))

df.boxplot(column = 'backers_count', by='reach_goal', ax = ax)
ax.set_yscale('log')
plt.show()
## It makes sense that there are less backer that reaching their goal isn't a possibility. 
```

```python
fig, ax = plt.subplots(figsize=(4, 6))

ax = df.groupby(['country'])['reach_goal'].sum().plot.bar()

plt.show()
## It looks like a good portion of successful kickstarters occur in the US
## but this doesn't tell us much since there is a higher amount of kickststarters occuring there. What is the 
## success rate like?
```

```python
## Get the aggregate sum and count
country_calc = df.groupby('country')['reach_goal'].agg(['sum','count'])
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax = (country_calc['sum'] / country_calc['count']).plot.bar()

plt.show()
## Right away, we see that success rate is the best in HK (Hong Kong) at 74.7% 
## with JP (Japan) trailing close behind at 67.7%. 
## The US is around 58% success rate. 
## IT (Italy) has the smallest success rate at 32.8%
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['goal'], df['percent_goal'], alpha = 0.1, s = 0.5)
ax.set_ylim([-2,600])
ax.set_xlim([0,2000000])

ax.set_xlabel('Goal')
ax.set_ylabel('Percent Reached of Goal')
ax.set_title('Goal Amount Success')

plt.show()

## Need to zoom in super close otherwise the extreme will distort the graph. 
## It looks like a pretty negative correlation as you increase in goal amount (which by logic makes sense)
## However, for amounts less than 250,000 it's harder to tell from this since
## there are a lot of points that run all over the place. 
## Setting the alpha to 0.1 and size to 0.5, there's a very intersting pattern that appears.
## It's almost like a second graph on top where it starts at 100%.
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['goal'], df['percent_goal'], alpha = 0.1, s = 0.5)
ax.set_ylim([-2,300])
ax.set_xlim([0,300000])

ax.set_xlabel('Goal')
ax.set_ylabel('Percent Reached of Goal')
ax.set_title('Goal Amount Success')

plt.show()
## For a lot of projects, they are either clearly hitting their goal or clearly not. 
## I.E. as we get closer to 250,000, there are less and less projects reaching around 50%
## but more jump to 100% showing that they're more likely to know DURING their kickstarter
## whether or not they'll succeed. If they've reached around 75%, there's a higher chance they'll finish 
## than they will not finish. 
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df['img_count'], df['percent_goal'], alpha = 0.3, s = 0.5)
ax.set_ylim([-5,500])

ax.set_xlabel('Amount of Images')
ax.set_ylabel('Percent Reached of Goal')
ax.set_title('Image Amount to Success')

plt.show()
## It looks like the amount of images portrays a similar phenomenon that goal and percentage did. 
## It's also easier to see a drop off that a higher amount of images doesn't necessarily mean
## a higher success rate.
```
